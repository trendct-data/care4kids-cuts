# Care4Kids enrollment

Checking out the claim that enrollment is dropping (25% July 2016 to 2017) since cuts to program.

```python
import pandas as pd
from calendar import month_name
```

```python
muni_col = "Municipality"
tot_col = "Total \n(This is a distinct count, not a sum of the columns)"

def get_data(month,year):
    
    fname = "data/" + month_name[month] + "-"+ str(year) +"-Excel.xls"
    
    try:
        df = pd.read_excel(fname,skiprows=10)
    except: 
        df = pd.read_excel(fname + "x",skiprows=10)
        
    return df[[muni_col,tot_col]]

get_data(1,2016)
```

```python
# get wealth rank of towns
def get_wealth_data():
    
    df = pd.read_excel("data/wealth rank.xls",skiprows=15)
    
    return df

def get_wealth(town):
    
    df = get_wealth_data()
    
    return df[df["Name"].str.strip() == str(town).upper().strip()]


def get_wealth_rank(town):
    ret = get_wealth(town)
    
    if len(ret.index) != 1: return None
    return int(ret.reset_index().iloc[0]["Rank"])

get_wealth_data()
get_wealth("Ansonia")
get_wealth_rank("Ansonia")
```

```python
def get_total(month,year):
    
    df = get_data(month,year)
    
    return int(df[df[muni_col] == "Total"][tot_col])

get_total(7,2017)
```

```python
def get_yoy():
    
    ret = []
    for y in [2015,2016,2017]:
        row = {"year":y}
        for m in range(1,12):
            tot = get_total(m,y)
            #print y, m, ":",tot
            row[m] = get_total(m,y)
        ret.append(row)
    return pd.DataFrame(ret)
            
get_yoy()
```

```python
# Check the 2016 to 2017 percentage for each month

month_diffs = get_yoy().set_index("year").transpose()
month_diffs["diff"] = month_diffs[2016] - month_diffs[2017]
month_diffs["diff_pct"] = month_diffs["diff"]/month_diffs[2016]

month_diffs
```

```python
def get_towns():
    
    frames = []
    
    for y in [2015,2016,2017]:
        for m in range(1,8):
            
            df = get_data(m,y).set_index(muni_col)
            new_col = str(y) + "-" + str(m)
            df[new_col] = df[tot_col]
            frames.append(df.reset_index()[[muni_col,new_col]])
                    
    ret = frames[0].set_index(muni_col)
    
    for df in frames[1:]:
        ret = ret.join(df.set_index(muni_col))

    # add wealth rank data
    ret["wealth rank"] = ret.index.map(get_wealth_rank)

    return ret

get_towns()
```

```python
def get_diff(month=7):
    
    cols = map(lambda x: str(x) + "-" + str(month),[2015,2016,2017]) + ["wealth rank"]
    
    df = get_towns()[cols]
    
    df["diff"] = df["2016-" + str(month)] - df["2017-" + str(month)]
    df["diff_pct"] = (df["diff"] * 100 /  df["2016-" + str(month)]).round()
    
    return df
    
get_diff().to_csv("output/diff-july-2016-2017.csv")
get_diff()
```
